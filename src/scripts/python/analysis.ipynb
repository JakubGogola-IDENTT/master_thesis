```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
solvers = ['Chuffed', 'CPOptimizer', 'OR-Tools', 'Gecode']
sizes = [5, 10, 15, 19, 23]
```

```python
summary_sync = pd.read_csv('summary_sync.csv')
summary_sync
```

```python
summary_parallel = pd.read_csv('summary_parallel.csv')
summary_parallel
```

```python
groupping_args = 'solver'

ss = summary_sync.groupby(groupping_args)
sp = summary_parallel.groupby(groupping_args)
```

```python
x = np.linspace(1, 5, 5)
fig, ax = plt.subplots(figsize=(15,10))


for solver in solvers:    
    gr = ss.get_group(solver)
    ax.plot(x, gr['steps'])
```
